```python
import pandas as pd
import json
```

```python
metadata = pd.read_csv("metadata.csv")
```

```python
metadata
```

```python
for i in range(len(metadata)):
    new_dict = dict()
    attr = []
    for key, val in dict(metadata.iloc[i]).items():
        val = str(val)
        if key.startswith("attributes"):
            attr_dict = {}
            attr_dict['trait_type'] = key[10:]
            attr_dict['value'] = val
            attr.append(attr_dict)
        else:
            new_dict[key] = val
    new_dict["attributes"] = attr
    
    with open('./'+str(i)+'.json', 'w') as fp:
        json.dump(new_dict, fp, indent=3)
```
